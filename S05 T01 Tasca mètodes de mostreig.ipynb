**Systematic Sampling**

```python
# Define systematic sampling function
def systematic_sampling(df, step):
    initial_value = np.random.randint(0,111)
    indexes = np.arange(initial_value, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
 
 
# Obtain a systematic sample and save it in a new variable
df_syst_sample = systematic_sampling(df_NBA, 625) # 625 is the step of the systematic sampling. We select it to have around 1000 players
```

**SMOTE Sample (Synthetic Minority Oversampling Technique)**

We are going to check the players that have play and they one that haven´t and use the SMOTE to simulate addional samples and avoid high differences in the variable weight.

```python
# !pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

SMOTE will only work with numerical values, so we have to clean up the dataset.

For this exercise we keep it simple and we remove string columns, fill all nulls and crease a column with a 1 if the players have played any minute or with a 0 if they haven´t played any minute.

```python
df_NBA['MIN'].fillna(0, inplace=True)
df_NBA.MIN.value_counts()
```

```python
df_NBA['played']=1
df_NBA.loc[(df_NBA['MIN']==0), 'played']=0
df_NBA['played'].value_counts()           
```

```python
# We remove string columns and fill all NaN with 0
df_NBA2 = df_NBA.copy()
df_NBA2.drop(['GAME_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'TEAM_CITY', 'PLAYER_ID',
       'PLAYER_NAME', 'NICKNAME', 'START_POSITION', 'COMMENT', 'MIN'], axis=1, inplace=True)
df_NBA2.fillna(0, inplace=True)
```

```python
x = df_NBA2.drop('played', axis=1) #x is the NBA data set with only numeric values but the played column
y = df_NBA2.played
print('Original dataset shape %s' % Counter(y))
```

523751 players have played some minute and 102360 players have NOT played any minute. There is a big difference in % so we will use SMOTE to simulate further players that don´t have played

```python
x
```

```python
smote = SMOTE()
x_sm, y_sm = smote.fit_resample(x, y)
print('Original dataset shape %s' % Counter(y_sm))
```

The number of players that have played or NOT any minutes is now the same

**Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.**

I love this problem statement:

Say you have a stream of items of large and unknown length that we can only iterate over once.
Create an algorithm that randomly chooses an item from this stream such that each item is equally likely to be selected.

We are going to select randomly 1000 players from our NBA Dataset

```python
import random
```

```python
k=1000
reservoir = []
for i, element in enumerate(df_NBA.PLAYER_ID):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            # Select item in df_NBA and remove one of the k items already selected
             reservoir[random.choice(range(0,k))] = element
```
